```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tic
from matplotlib.ticker import AutoMinorLocator
```

```python
distribution = 'lognormal'

plt.style.use('stylesheet.mpltstyle_cond_plots')
fig, ax = plt.subplots(2,2, figsize=(4.3,3.5), sharex = True)

CMoment = np.loadtxt('../Results/'+distribution+'_CM.txt')
unc = np.loadtxt('../Results/'+distribution+'_SM.txt')[0]*np.ones(100)

ax[0,0].set_ylabel('Conditional Mean')
ax[0,0].annotate("(a)", xy=(0.9, 0.9), xycoords="axes fraction")
ax[0,0].set_xlim(0,1)
ax[0,0].set_ylim(1E-9,1E-6)

ax[0,0].plot(np.linspace(0,1,100), unc, '--', linewidth = 1, c = 'gray', label = 'Unc. Mean')
ax[0,0].plot(np.linspace(0,1,100), CMoment[0,:], '.', c = 'black', label = r'$r_o$')
ax[0,0].plot(np.linspace(0,1,100), CMoment[1,:], '.', c = 'purple', label = r'$\phi_o$')
ax[0,0].plot(np.linspace(0,1,100), CMoment[2,:], '.', c = 'red', label = r'$p$')
ax[0,0].plot(np.linspace(0,1,100), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[0,0].plot(np.linspace(0,1,100), CMoment[4,:], '.', c = 'green', label = r'$T$')
ax[0,0].set_yscale('log')
ax[0,0].legend(ncol = 2, loc = 2)


CMoment = np.loadtxt('../Results/'+distribution+'_CV.txt')
unc = np.loadtxt('../Results/'+distribution+'_SM.txt')[1]*np.ones(100)

ax[0,1].set_ylabel('Conditional Variance')
ax[0,1].annotate("(b)", xy=(0.9, 0.9), xycoords="axes fraction")
ax[0,1].set_xlim(0,1)
ax[0,1].set_ylim(1E-18,1E-13)


ax[0,1].plot(np.linspace(0,1,100), unc, '--', linewidth = 1, c = 'gray', label = 'Unc. Variance')
ax[0,1].plot(np.linspace(0,1,100), CMoment[0,:], '.', c = 'black', label = r'$r_o$')
ax[0,1].plot(np.linspace(0,1,100), CMoment[1,:], '.', c = 'purple', label = r'$\phi_o$')
ax[0,1].plot(np.linspace(0,1,100), CMoment[2,:], '.', c = 'red', label = r'$p$')
ax[0,1].plot(np.linspace(0,1,100), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[0,1].plot(np.linspace(0,1,100), CMoment[4,:], '.', c = 'green', label = r'$T$')
ax[0,1].set_yscale('log')
ax[0,1].legend(ncol = 2)


CMoment = np.loadtxt('../Results/'+distribution+'_CS.txt')
unc = np.loadtxt('../Results/'+distribution+'_SM.txt')[2]*np.ones(100)

ax[1,0].set_xlabel('Normalized parameter')
ax[1,0].set_ylabel('Conditional Skewness')
ax[1,0].annotate("(c)", xy=(0.9, 0.9), xycoords="axes fraction")
ax[1,0].set_xlim(0,1)
ax[1,0].set_ylim(0,9)

ax[1,0].plot(np.linspace(0,1,100), unc, '--', linewidth = 1, c = 'gray', label = 'Unc. Skewness')
ax[1,0].plot(np.linspace(0,1,100), CMoment[0,:], '.', c = 'black', label = r'$r_o$')
ax[1,0].plot(np.linspace(0,1,100), CMoment[1,:], '.', c = 'purple', label = r'$\phi_o$')
ax[1,0].plot(np.linspace(0,1,100), CMoment[2,:], '.', c = 'red', label = r'$p$')
ax[1,0].plot(np.linspace(0,1,100), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[1,0].plot(np.linspace(0,1,100), CMoment[4,:], '.', c = 'green', label = r'$T$')
ax[1,0].legend(ncol = 2, loc = 2)


CMoment = np.loadtxt('../Results/'+distribution+'_CK.txt')
unc = np.loadtxt('../Results/'+distribution+'_SM.txt')[3]*np.ones(100)

ax[1,1].set_xlabel('Normalized parameter')
ax[1,1].set_ylabel('Conditional Kurtosis')
ax[1,1].annotate("(d)", xy=(0.9, 0.1), xycoords="axes fraction")
ax[1,1].set_xlim(0,1)
ax[1,1].set_ylim(1,1E2)

ax[1,1].plot(np.linspace(0,1,100), unc, '--', linewidth = 1, c = 'gray', label = 'Unc. Kurtosis')
ax[1,1].plot(np.linspace(0,1,100), CMoment[0,:], '.', c = 'black', label = r'$r_o$')
ax[1,1].plot(np.linspace(0,1,100), CMoment[1,:], '.', c = 'purple', label = r'$\phi_o$')
ax[1,1].plot(np.linspace(0,1,100), CMoment[2,:], '.', c = 'red', label = r'$p$')
ax[1,1].plot(np.linspace(0,1,100), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[1,1].plot(np.linspace(0,1,100), CMoment[4,:], '.', c = 'green', label = r'$T$')
ax[1,1].set_yscale('log')
ax[1,1].legend(ncol = 2)

plt.savefig('../Figures/cond_moments_'+distribution+'.pdf')
plt.show()
```
