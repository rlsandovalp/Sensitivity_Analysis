```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
```

```python
distribution = 'lognormal'
fig, ax = plt.subplots(2,2, figsize=(12,8))
size_label = 15
CMoment = np.loadtxt('../Results/'+distribution+'_CM.txt')
unc = np.loadtxt('../Results/'+distribution+'_SM.txt')[0]*np.ones(100)
ax[0,0].plot(np.linspace(0,1,100), unc, linewidth = 3, c = 'gray', label = 'Unc. Mean')
ax[0,0].text(0.9, 0.9, '(a)', transform = ax[0,0].transAxes, fontsize = size_label)
ax[0,0].plot(np.linspace(0,1,100), CMoment[0,:], '.', c = 'black', label = r'$r_o$')
ax[0,0].plot(np.linspace(0,1,100), CMoment[1,:], '.', c = 'purple', label = r'$\phi_o$')
ax[0,0].plot(np.linspace(0,1,100), CMoment[2,:], '.', c = 'red', label = r'$p$')
ax[0,0].plot(np.linspace(0,1,100), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[0,0].plot(np.linspace(0,1,100), CMoment[4,:], '.', c = 'green', label = r'$T$')
ax[0,0].set_ylim(1E-9,1E-6)
ax[0,0].set_yscale('log')
ax[0,0].set_ylabel('Conditional Mean', fontsize = size_label)
ax[0,0].tick_params(axis = "y", which = 'both', right = True, direction = "in", labelsize = size_label)
ax[0,0].xaxis.set_minor_locator(AutoMinorLocator())
ax[0,0].tick_params(axis = "x", which = 'both', top = True, direction = "in", labelsize = size_label)
ax[0,0].legend()

CMoment = np.loadtxt('../Results/'+distribution+'_CV.txt')
unc = np.loadtxt('../Results/'+distribution+'_SM.txt')[1]*np.ones(100)
ax[0,1].text(0.9, 0.9, '(b)', transform = ax[0,1].transAxes, fontsize = size_label)
ax[0,1].plot(np.linspace(0,1,100), unc, linewidth = 3, c = 'gray', label = 'Unc. Variance')
ax[0,1].plot(np.linspace(0,1,100), CMoment[0,:], '.', c = 'black', label = r'$r_o$')
ax[0,1].plot(np.linspace(0,1,100), CMoment[1,:], '.', c = 'purple', label = r'$\phi_o$')
ax[0,1].plot(np.linspace(0,1,100), CMoment[2,:], '.', c = 'red', label = r'$p$')
ax[0,1].plot(np.linspace(0,1,100), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[0,1].plot(np.linspace(0,1,100), CMoment[4,:], '.', c = 'green', label = r'$T$')
ax[0,1].set_ylim(1E-18,1E-13)
ax[0,1].set_yscale('log')
ax[0,1].set_ylabel('Conditional Variance', fontsize = size_label)
ax[0,1].tick_params(axis = "y", which = 'both', right = True, direction = "in", labelsize = size_label)
ax[0,1].xaxis.set_minor_locator(AutoMinorLocator())
ax[0,1].tick_params(axis = "x", which = 'both', top = True, direction = "in", labelsize = size_label)
ax[0,1].legend()

CMoment = np.loadtxt('../Results/'+distribution+'_CS.txt')
unc = np.loadtxt('../Results/'+distribution+'_SM.txt')[2]*np.ones(100)
ax[1,0].text(0.9, 0.9, '(c)', transform = ax[1,0].transAxes, fontsize = size_label)
ax[1,0].plot(np.linspace(0,1,100), unc, linewidth = 3, c = 'gray', label = 'Unc. Skewness')
ax[1,0].plot(np.linspace(0,1,100), CMoment[0,:], '.', c = 'black', label = r'$r_o$')
ax[1,0].plot(np.linspace(0,1,100), CMoment[1,:], '.', c = 'purple', label = r'$\phi_o$')
ax[1,0].plot(np.linspace(0,1,100), CMoment[2,:], '.', c = 'red', label = r'$p$')
ax[1,0].plot(np.linspace(0,1,100), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[1,0].plot(np.linspace(0,1,100), CMoment[4,:], '.', c = 'green', label = r'$T$')
ax[1,0].set_ylim(0,8)
ax[1,0].set_xlabel('Normalized parameter', fontsize = size_label)
ax[1,0].set_ylabel('Conditional Skewness', fontsize = size_label)
ax[1,0].tick_params(axis = "y", which = 'both', right = True, direction = "in", labelsize = size_label)
ax[1,0].xaxis.set_minor_locator(AutoMinorLocator())
ax[1,0].tick_params(axis = "x", which = 'both', top = True, direction = "in", labelsize = size_label)
ax[1,0].legend()

CMoment = np.loadtxt('../Results/'+distribution+'_CK.txt')
unc = np.loadtxt('../Results/'+distribution+'_SM.txt')[3]*np.ones(100)
ax[1,1].text(0.9, 0.9, '(d)', transform = ax[1,1].transAxes, fontsize = size_label)
ax[1,1].plot(np.linspace(0,1,100), unc, linewidth = 3, c = 'gray', label = 'Unc. Kurtosis')
ax[1,1].plot(np.linspace(0,1,100), CMoment[0,:], '.', c = 'black', label = r'$r_o$')
ax[1,1].plot(np.linspace(0,1,100), CMoment[1,:], '.', c = 'purple', label = r'$\phi_o$')
ax[1,1].plot(np.linspace(0,1,100), CMoment[2,:], '.', c = 'red', label = r'$p$')
ax[1,1].plot(np.linspace(0,1,100), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[1,1].plot(np.linspace(0,1,100), CMoment[4,:], '.', c = 'green', label = r'$T$')
ax[1,1].set_ylim(1,1E2)
ax[1,1].set_yscale('log')
ax[1,1].set_xlabel('Normalized parameter', fontsize = size_label)
ax[1,1].set_ylabel('Conditional Kurtosis', fontsize = size_label)
ax[1,1].tick_params(axis = "y", which = 'both', right = True, direction = "in", labelsize = size_label)
ax[1,1].xaxis.set_minor_locator(AutoMinorLocator())
ax[1,1].tick_params(axis = "x", which = 'both', top = True, direction = "in", labelsize = size_label)
ax[1,1].legend()

plt.tight_layout()
plt.savefig('../Figures/cond_moments_'+distribution+'.pdf')
plt.show()
```
