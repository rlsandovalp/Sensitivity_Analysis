```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
output = 'J'
case = 'case_b'

plt.style.use('stylesheet.mpltstyle_cond_plots')
fig, ax = plt.subplots(2,2, figsize=(4.3,3.5), sharex = True)

CMoment = np.loadtxt('../Results/SA/'+output+'/'+case+'/CM.txt')
classes = np.shape(CMoment)[1]
unc = np.loadtxt('../Results/SA/'+output+'/'+case+'/SM.txt')[0]*np.ones(classes)

ax[0,0].set_ylabel('Conditional Mean')
ax[0,0].annotate("(a)", xy=(0.9, 0.9), xycoords="axes fraction")
ax[0,0].set_xlim(0,1)
ax[0,0].set_ylim(1E-9,1E-6)

ax[0,0].plot(np.linspace(0,1,classes), unc, '--', linewidth = 1, c = 'gray', label = 'Unc.')
ax[0,0].plot(np.linspace(0,1,classes), CMoment[1,:], '.', c = 'black', label = r'$r_o$')
ax[0,0].plot(np.linspace(0,1,classes), CMoment[2,:], '.', c = 'purple', label = r'$\phi_o$')
ax[0,0].plot(np.linspace(0,1,classes), CMoment[0,:], '.', c = 'red', label = r'$p$')
ax[0,0].plot(np.linspace(0,1,classes), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[0,0].plot(np.linspace(0,1,classes), CMoment[14,:], '.', c = 'green', label = r'$T$')
ax[0,0].set_yscale('log')
ax[0,0].legend(ncol = 2, loc = 2)


CMoment = np.loadtxt('../Results/SA/'+output+'/'+case+'/CV.txt')
unc = np.loadtxt('../Results/SA/'+output+'/'+case+'/SM.txt')[1]*np.ones(classes)

ax[0,1].set_ylabel('Conditional Variance')
ax[0,1].annotate("(b)", xy=(0.9, 0.9), xycoords="axes fraction")
ax[0,1].set_xlim(0,1)
ax[0,1].set_ylim(1E-18,1E-13)


ax[0,1].plot(np.linspace(0,1,classes), unc, '--', linewidth = 1, c = 'gray', label = 'Unc.')
ax[0,1].plot(np.linspace(0,1,classes), CMoment[1,:], '.', c = 'black', label = r'$r_o$')
ax[0,1].plot(np.linspace(0,1,classes), CMoment[2,:], '.', c = 'purple', label = r'$\phi_o$')
ax[0,1].plot(np.linspace(0,1,classes), CMoment[0,:], '.', c = 'red', label = r'$p$')
ax[0,1].plot(np.linspace(0,1,classes), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[0,1].plot(np.linspace(0,1,classes), CMoment[14,:], '.', c = 'green', label = r'$T$')
ax[0,1].set_yscale('log')
ax[0,1].legend(ncol = 2, loc = 4)


CMoment = np.loadtxt('../Results/SA/'+output+'/'+case+'/CS.txt')
unc = np.loadtxt('../Results/SA/'+output+'/'+case+'/SM.txt')[2]*np.ones(classes)

ax[1,0].set_xlabel('Normalized parameter')
ax[1,0].set_ylabel('Conditional Skewness')
ax[1,0].annotate("(c)", xy=(0.05, 0.9), xycoords="axes fraction")
ax[1,0].set_xlim(0,1)
ax[1,0].set_ylim(0,9)

ax[1,0].plot(np.linspace(0,1,classes), unc, '--', linewidth = 1, c = 'gray', label = 'Unc.')
ax[1,0].plot(np.linspace(0,1,classes), CMoment[1,:], '.', c = 'black', label = r'$r_o$')
ax[1,0].plot(np.linspace(0,1,classes), CMoment[2,:], '.', c = 'purple', label = r'$\phi_o$')
ax[1,0].plot(np.linspace(0,1,classes), CMoment[0,:], '.', c = 'red', label = r'$p$')
ax[1,0].plot(np.linspace(0,1,classes), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[1,0].plot(np.linspace(0,1,classes), CMoment[14,:], '.', c = 'green', label = r'$T$')
ax[1,0].legend(ncol = 2, loc = 1)


CMoment = np.loadtxt('../Results/SA/'+output+'/'+case+'/CK.txt')
unc = np.loadtxt('../Results/SA/'+output+'/'+case+'/SM.txt')[3]*np.ones(classes)

ax[1,1].set_xlabel('Normalized parameter')
ax[1,1].set_ylabel('Conditional Kurtosis')
ax[1,1].annotate("(d)", xy=(0.9, 0.1), xycoords="axes fraction")
ax[1,1].set_xlim(0,1)
ax[1,1].set_ylim(1e-0,1e2)

ax[1,1].plot(np.linspace(0,1,classes), unc, '--', linewidth = 1, c = 'gray', label = 'Unc.')
ax[1,1].plot(np.linspace(0,1,classes), CMoment[1,:], '.', c = 'black', label = r'$r_o$')
ax[1,1].plot(np.linspace(0,1,classes), CMoment[2,:], '.', c = 'purple', label = r'$\phi_o$')
ax[1,1].plot(np.linspace(0,1,classes), CMoment[0,:], '.', c = 'red', label = r'$p$')
ax[1,1].plot(np.linspace(0,1,classes), CMoment[3,:], '.', c = 'blue', label = r'$\tau$')
ax[1,1].plot(np.linspace(0,1,classes), CMoment[14,:], '.', c = 'green', label = r'$T$')
ax[1,1].set_yscale('log')
ax[1,1].legend(ncol = 2)

plt.savefig('../Figures/cond_moments_'+output+'_'+case+'.pdf')
plt.show()
```
